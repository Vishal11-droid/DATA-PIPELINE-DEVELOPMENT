```python
import pandas as pd
from sklearn.datasets import load_iris

# Loading iris data....
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Adding a target column...
df['target'] = data.target

#viewing the dataframe....
df.head()
```

```python
df.isnull().sum()  # for finding the missing values in the dataset...
```

```python
#feature scaling....
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop('target', axis=1))

processed_df = pd.DataFrame(scaled_features, columns=data.feature_names)
processed_df['target'] = df['target']

processed_df.head()
```

```python
# saving our preprocessed data as csv :)
processed_df.to_csv('processed_iris.csv', index=False)
```
